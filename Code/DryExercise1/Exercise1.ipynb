```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
R = 1
V = [1, 2*R]
for d in range(2,11):
    V.append((2*np.pi/d)*(R**2)*V[d-2])
print(V)
```

```python
eps=[0.2, 0.1, 0.01]
data={0.2:[], 0.1:[], 0.01:[]}
for e in eps:
    R_new = R - e
    V_new = [1, 2*R_new]
    for d in range(2,11):
        V_new.append((2*np.pi/d)*(R_new**2)*V_new[d-2])
    eta = [(V[i]-V_new[i])/V[i] for i in range(2,len(V))]
    data[e] = eta
print(data)
```

```python
d = range(2,11)
plt.plot(d, data[0.2], color='r', label='$\epsilon=0.2$')
plt.plot(d, data[0.1], color='g', label='$\epsilon=0.1$')
plt.plot(d, data[0.01], color='b', label='$\epsilon=0.01$')
  
# Naming the x-axis, y-axis and the whole graph
plt.xlabel("Dimension (d)")
plt.ylabel("$\eta_d (\epsilon)$")
plt.title("Volume Fraction that's Distance $\epsilon$ from Surface of $d$-dimensional Unit Ball as a function of $d$")
  
plt.legend()
plt.show()
```
